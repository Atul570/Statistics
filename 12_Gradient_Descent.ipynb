## <font color = "red">üîÅ Gradient Descent </font>

Gradient Descent is a widely-used optimization algorithm in machine learning and deep learning. It aims to minimize the cost function, which measures the error between the predicted and actual results. By iteratively adjusting the model's parameters (weights and biases) in the direction of the negative gradient of the cost function, Gradient Descent helps the model converge to the optimal parameter values.

**üéØ Goal:**
To find the value of `b` (intercept) that minimizes the **Loss Function**  
for the linear regression equation:

$$
\hat{y}_i = mx_i + b
$$

We assume **`m` is constant** and only optimize **`b`** using **Gradient Descent**.



**üßÆ 1. Loss Function (Mean Squared Error)**

$$
L = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - (mx_i + b))^2
$$

Our objective is to minimize `L` w.r.t. `b`.



**üîç 2. Derivative of Loss Function w.r.t. `b`**

Using chain rule:

$$
\frac{dL}{db} = \frac{d}{db} \sum_{i=1}^{n} (y_i - mx_i - b)^2
$$

Now apply the derivative:

$$
\frac{dL}{db} = 2 \sum_{i=1}^{n} (y_i - mx_i - b)(-1)
$$

$$
\Rightarrow \frac{dL}{db} = -2 \sum_{i=1}^{n} (y_i - mx_i - b)
$$

So the **slope** (i.e., gradient) is:

$$
\text{slope}_b = -2 \sum_{i=1}^{n} (y_i - mx_i - b)
$$

This slope tells us the **direction and magnitude** of change required in `b`.

**üìä What This Graph Shows:**

* The blue curve is the **loss function**:

  $$
  L = (b - 3)^2
  $$

  This mimics the squared loss you'd minimize in linear regression.

* The **minimum** of the loss occurs at **b = 3**, where slope = 0.

* The **red arrows** at different points (b = -6, 0, 6) show the **gradient (slope)**:

  * At **b = -6**, slope is **negative**, so gradient descent pushes `b` **rightward**
  * At **b = 6**, slope is **positive**, so gradient descent pushes `b` **leftward**
  * At **b = 0**, slope is smaller (still negative), so a smaller step to the right

**üìò Connect It to Your Derivation:**

* This parabola represents:

  $$
  L(b) = \sum (y_i - mx_i - b)^2
  $$
* Your derivation showed:

  $$
  \frac{dL}{db} = -2 \sum (y_i - mx_i - b)
  $$

  Which plays the role of the slope here.
* Gradient Descent updates:

  $$
  b = b - \eta \cdot \text{slope}
  $$


**üß† 3. Gradient Descent Update Rule**

We update `b` using the formula:

$$
b_{\text{new}} = b_{\text{old}} - \eta \cdot \text{slope}_b
$$

Where:
- `Œ∑` (eta) is the **learning rate** ‚Äì it controls the step size
- `slope_b` is the gradient we computed above



**üîÅ 4. Iteration Loop**

We repeat this update inside a loop (called **epochs**), until we reach **convergence**:

```python
for i in range(epochs):
    b_new = b_old - Œ∑ * slope_b
````

You can initialize `b = 0`, and iterate many times, updating it at each step.



**üõë 5. When to Stop Iteration?**

We stop when:

* The **change in loss is very small** (convergence)
* OR the **number of epochs** is reached
* OR the **change in `b` is very small**, i.e.

  $$
  |b_{\text{new}} - b_{\text{old}}| < \text{tolerance}
  $$



**‚ö†Ô∏è 6. Role of Learning Rate `Œ∑`**

* If `Œ∑` is **too small** ‚Üí slow learning, may take too many steps
* If `Œ∑` is **too large** ‚Üí may overshoot the minimum and diverge

A good value is often found by **trial & error** (like `0.01`, `0.001` etc.)



‚úÖ Summary

* We fix `m` and optimize `b` using gradient descent
* We derived slope using the chain rule
* Update rule:

  $$
  b = b - \eta \cdot (-2 \sum (y - mx - b))
  $$
* Repeat for multiple epochs to reduce loss
* Stop when convergence is reached

> üìå This logic is extendable to optimize both `m` and `b` together as the next step in full linear regression training.

## <font color="red">üîÅ Gradient Descent (Optimize Both `m` and `b`)</font>

Gradient Descent helps us **minimize the cost function** by iteratively updating **both parameters**:

* `m` ‚Üí the slope of the regression line
* `b` ‚Üí the intercept of the regression line

We‚Äôll work with the **linear regression equation**:

$$
\hat{y}_i = mx_i + b
$$



**üéØ Goal:**

To find the values of **`m`** and **`b`** that **minimize the Loss Function (MSE)**:

$$
L = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - (mx_i + b))^2
$$



**üßÆ 1. Loss Function: Mean Squared Error**

We define:

$$
L(m, b) = \sum_{i=1}^{n} (y_i - (mx_i + b))^2
$$

This measures the **total squared error** between actual values `y_i` and predicted values `\hat{y}_i`.



**üîç 2. Partial Derivatives**

We now take **partial derivatives of the loss function** with respect to both `m` and `b` to compute the **gradients (slopes)**.



**üî∫ Partial Derivative w\.r.t. `m`:**

$$
\frac{\partial L}{\partial m} = \frac{d}{dm} \sum (y_i - (mx_i + b))^2
$$

Apply chain rule:

$$
= \sum 2(y_i - mx_i - b)(-x_i) = -2 \sum x_i (y_i - mx_i - b)
$$

So,

$$
\text{slope}_m = -2 \sum x_i (y_i - (mx_i + b))
$$



**üîª Partial Derivative w\.r.t. `b`:**

$$
\frac{\partial L}{\partial b} = \frac{d}{db} \sum (y_i - (mx_i + b))^2
$$

$$
= \sum 2(y_i - mx_i - b)(-1) = -2 \sum (y_i - mx_i - b)
$$

So,

$$
\text{slope}_b = -2 \sum (y_i - (mx_i + b))
$$



**üîÅ 3. Gradient Descent Update Rules**

We now update both `m` and `b` using the slopes calculated above.

**General form:**

$$
\begin{aligned}
m_{\text{new}} &= m_{\text{old}} - \eta \cdot \text{slope}_m \\
b_{\text{new}} &= b_{\text{old}} - \eta \cdot \text{slope}_b
\end{aligned}
$$

Where:

* `Œ∑` (eta) is the **learning rate** (e.g., 0.01, 0.001)
* `slope_m` and `slope_b` are the gradients we calculated



## üîÅ 4. Full Iteration Loop (Epochs)

```python
# Pseudocode:
for epoch in range(num_epochs):
    y_hat = m * X + b
    error = y - y_hat

    slope_m = -2 * sum(X * error)
    slope_b = -2 * sum(error)

    m = m - eta * slope_m
    b = b - eta * slope_b
```

Each iteration updates `m` and `b` to reduce the error.



**üõë 5. When to Stop Iterating?**

Stop gradient descent when:

* Change in **loss is very small** (convergence)
* A **maximum number of epochs** is reached
* Change in both `m` and `b` is **less than a threshold**:

$$
|m_{\text{new}} - m_{\text{old}}| < \epsilon, \quad |b_{\text{new}} - b_{\text{old}}| < \epsilon
$$



**‚ö†Ô∏è 6. Role of Learning Rate `Œ∑`**


* If `Œ∑` is **too small**: training is **very slow**
* If `Œ∑` is **too large**: may **overshoot** or **diverge**

### Effect of loss Funtion (Convex Function and Non-convex Function)

üîç **Quick Background First**

In machine learning, we use a **loss function** to measure how well our model is doing. Our goal is to **minimize** this function.

Depending on the **shape** of the loss function, optimization behaves differently:

| Type                    | Example                      | Optimization Goal                                      |
| ----------------------- | ---------------------------- | ------------------------------------------------------ |
| **Convex Function**     | Linear Regression (MSE Loss) | Easy to find **global minimum**                        |
| **Non-Convex Function** | Deep Neural Networks         | May get stuck in **local minima** or **saddle points** |



üß† **Key Concepts You Need to Know**

1. **Loss Function**

Gives us a number that shows how far the model is from the correct answer. We want to **minimize** it.

2. **Convex Function**

Looks like a smooth U-shaped curve or bowl. Has **only one** lowest point (global minimum).

3. **Non-Convex Function**

Can look like hills and valleys. It has **many local minima**, and **only one** of them is the **global minimum**.

4. **Local Minimum vs Global Minimum**

* **Local Minimum**: A point lower than neighbors but not the lowest possible.
* **Global Minimum**: The **absolute lowest** point in the entire function.


üé• Animation-style Explanation (Think Visually)

I‚Äôll explain this in a way that lets you imagine it like an animation.


‚úÖ Convex Loss Function ‚Äì Animation Mental Model:

**Imagine this:**

1. A smooth **U-shaped bowl**.
2. A ball is dropped anywhere on the bowl.
3. It **rolls down** smoothly to the **center**, which is the **global minimum**.

**Why this works:**

* In convex functions, the gradient (slope) always points toward the global minimum.
* Gradient Descent is guaranteed to find that global minimum.

üìå **Example in real life**: Linear regression using MSE loss.



‚ùå Non-Convex Loss Function ‚Äì Animation Mental Model:

**Now imagine this:**

1. A **bumpy landscape** with many **hills and valleys**.
2. A ball is dropped randomly on the landscape.
3. It **might settle in a nearby valley**, thinking it's the lowest point (local minimum).
4. But elsewhere, there's a **deeper valley** (global minimum), which it **never reaches** unless it‚Äôs lucky.

üìå **Real life**: Deep learning models with millions of weights form **non-convex** loss surfaces.



üìâ With Gradient Descent

Let‚Äôs say you‚Äôre using **gradient descent** to optimize your model:

* In **convex**: It‚Äôs like having a map with one destination. You‚Äôll always get there.
* In **non-convex**: It‚Äôs like a jungle. You might stop at a cave thinking it‚Äôs the end, but the treasure is still farther.


üîß Techniques to Deal With Non-Convexity

* **Momentum**: Helps escape shallow valleys.
* **Adam Optimizer**: Uses adaptive learning rates to navigate better.
* **Learning Rate Schedules**: Reduce LR gradually to fine-tune.
* **Stochasticity (SGD)**: Randomness helps in escaping local minima.
* **Ensemble Training / Restarts**: Try optimization multiple times from different starting points.


‚úÖ Summary Table

| Feature          | Convex Function             | Non-Convex Function             |
| ---------------- | --------------------------- | ------------------------------- |
| Shape            | Smooth bowl (U-shape)       | Bumpy landscape                 |
| Local Minima     | None                        | Many local minima               |
| Global Minima    | One, easy to reach          | One, hard to guarantee          |
| Gradient Descent | Always finds global minimum | Might get stuck in local minima |
| Example          | Linear Regression           | Neural Networks                 |



`````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation, PillowWriter

# Define a concave (inverted bowl) function with a clear local maximum
def f(x, y):
    return -((x - 2)**2 + (y - 2)**2)

# Its gradient (partial derivatives)
def grad(x, y):
    dx = -2 * (x - 2)
    dy = -2 * (y - 2)
    return dx, dy

# Gradient "descent" will now climb to the peak due to negative slope
start = np.array([-2.0, -2.0])  # start far from the peak
lr = 0.1
steps = 50
path = [start]

for _ in range(steps):
    x, y = path[-1]
    dx, dy = grad(x, y)
    new_point = np.array([x - lr * dx, y - lr * dy])
    path.append(new_point)

path = np.array(path)
z_path = f(path[:, 0], path[:, 1])

# Prepare meshgrid for surface
x = np.linspace(-4, 8, 300)
y = np.linspace(-4, 8, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 3D plot setup
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.75, edgecolor='none')

# Descent path
point, = ax.plot([], [], [], 'ro', markersize=6)
line, = ax.plot([], [], [], 'r-', linewidth=2)

def update(i):
    point.set_data(path[i, 0], path[i, 1])
    point.set_3d_properties(z_path[i])
    line.set_data(path[:i+1, 0], path[:i+1, 1])
    line.set_3d_properties(z_path[:i+1])
    return point, line

# Animate and save as .gif
anim = FuncAnimation(fig, update, frames=len(path), interval=100, blit=False)
anim.save("gradient_descent_local_maximum.gif", writer=PillowWriter(fps=10))

plt.title("Gradient Descent Reaching a Local Maximum")
plt.show()
`````

### <font color = "red">**Gradient Descent For Multiple Regression Loss Fnction**</font>

**üöÄ Gradient Descent Derivation for Multiple Linear Regression (Feature-wise)**



**üßÆ Step 1: The Model (Hypothesis Function)**

Suppose we have:

* `n` training samples
* `k` input features (independent variables)

**Model equation for the i-th sample:**

$$
\hat{y}_i = \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_k x_{ik} + b
$$

Or in summation:

$$
\hat{y}_i = \sum_{j=1}^{k} \beta_j x_{ij} + b
$$

Where:

* $x_{ij}$ is the **j-th feature value** for the **i-th data point**
* $\beta_j$ is the coefficient (slope) for feature j
* $b$ is the bias (intercept)
* $\hat{y}_i$ is the predicted value


**üéØ Step 2: The Loss Function (Mean Squared Error)**

We use MSE over all n samples:

$$
L = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2 = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \left(\sum_{j=1}^{k} \beta_j x_{ij} + b\right) \right)^2
$$

Our goal is to minimize `L` with respect to each `Œ≤_j` and `b`.



**üîç Step 3: Derivatives ‚Äî One by One (Start with Œ≤‚ÇÅ)**

**Let‚Äôs find derivative of L with respect to **Œ≤‚ÇÅ**:**

$$
\frac{\partial L}{\partial \beta_1} = \frac{\partial}{\partial \beta_1} \left[ \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \left( \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_k x_{ik} + b \right) \right)^2 \right]
$$

Only **Œ≤‚ÇÅ** is variable here, others are constant in this partial derivative.

$$
\frac{\partial L}{\partial \beta_1} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) x_{i1}
$$


**‚úÖ Similarly, for Œ≤‚ÇÇ:**

$$
\frac{\partial L}{\partial \beta_2} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) x_{i2}
$$



**‚úÖ For general Œ≤‚±º:**

$$
\frac{\partial L}{\partial \beta_j} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) x_{ij}
$$

That‚Äôs the **gradient for Œ≤‚±º**.


**üîÅ Step 4: Derivative w\.r.t. Bias `b`**

Bias doesn‚Äôt multiply with any feature ‚Äî it's added directly ‚Äî so:

$$
\frac{\partial L}{\partial b} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)
$$


**üõ†Ô∏è Step 5: Gradient Descent Update Rules**

We update each coefficient and bias like this:

$$
\beta_j \leftarrow \beta_j - \eta \cdot \frac{\partial L}{\partial \beta_j}
$$

$$
b \leftarrow b - \eta \cdot \frac{\partial L}{\partial b}
$$

Where:

* `Œ∑` is the learning rate
* Gradients we derived above go in place of `‚àÇL/‚àÇŒ≤_j` and `‚àÇL/‚àÇb`




**‚úÖ Summary of Final Gradient Formulas**

| Parameter       | Gradient Expression                                    |
| --------------- | ------------------------------------------------------ |
| Œ≤‚±º (j-th coeff) | $\frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) x_{ij}$ |
| b (bias)        | $\frac{-2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)$        |


**üìò Recap With Real Meaning**

For a model like:

$$
\hat{y} = \beta_1 \cdot \text{age} + \beta_2 \cdot \text{income} + \beta_3 \cdot \text{score} + b
$$

We are adjusting:

* Œ≤‚ÇÅ based on how error correlates with `age`
* Œ≤‚ÇÇ based on how error correlates with `income`
* Œ≤‚ÇÉ based on how error correlates with `score`
* `b` based on the average error



**üìò Example:**

Let‚Äôs say:

* You have 3 features: `age`, `income`, `education_score`
* Output: `spending_score`

Then the model is:

$$
\hat{y} = w_1 x_{\text{age}} + w_2 x_{\text{income}} + w_3 x_{\text{edu}} + b
$$

You plug in all data, compute prediction, error, gradient w\.r.t. each `w` and `b`, update them ‚Üí repeat!
