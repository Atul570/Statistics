```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
# Genrating poupulation :- 
mu = 5
sigma = 0.4
population = np.random.lognormal(mean=mu , sigma= sigma , size = 300000)

sns.kdeplot(x=population)
plt.plot()
```

```python
#creating sampling Distribution of mean = 
sampling_dist_mean = []
for i in range(0,100):
    sampling_dist_mean.append((np.random.choice(population,size = 15)).mean())
```

```python
fig ,ax = plt.subplots(2,2,figsize= (20,10))
sns.histplot(x=population,ax=ax[0][0])
ax[0][0].set_title("Histogram of Population")
sns.kdeplot(x=population ,ax=ax[0][1])
ax[0][1].set_title("Pdf of Population")
sns.histplot(x=sampling_dist_mean , ax = ax[1][0])
ax[1][0].set_title("Histogram of (sampling distribution of mean)")
sns.kdeplot(x=sampling_dist_mean,ax=ax[1][1])
ax[1][1].set_title("Pdf of (sampling distribution of mean)")
plt.plot()
```

```python
print("Mean of Population=",population.mean())
print("Standard Devation of Population = ",population.std())
print("Mean of (Sampling Distribution of mean)=",pd.Series(sampling_dist_mean).mean())
print("Standard Devation of (Sampling Distribution of mean)n = ",pd.Series(sampling_dist_mean).std())
print("-----------------------------------------------")
print("Geting Idea of standard devation  of population from sampling distribution of mean :- ")
print(pd.Series(sampling_dist_mean).std()*math.sqrt(15))
```
