<font color = "red" size = 12>**Hypothesis Testing**</font>

Hypothesis testing is a statistical method that is used to make a statistical decision using experimental data. Hypothesis testing is basically an assumption that we make about a population parameter. It evaluates two mutually exclusive statements about a population to determine which statement is best supported by the sample data. 

A statistical hypothesis test is a method of statistical inference used to decide whether the data at hand sufficiently support a particular hypothesis. Hypothesis testing allows us to make probabilistic statements about population parameters.

In hypothesis testing, an analyst tests a statistical sample, intending to provide evidence on the plausibility of the null hypothesis. Statistical analysts measure and examine a random sample of the population being analyzed. All analysts use a random population sample to test two different hypotheses: the null hypothesis and the alternative hypothesis.

<font color = "red">**Null and Alternative Hypotheses:-**</font>

The actual test begins by considering two hypotheses. They are called the null hypothesis and the alternative hypothesis. These hypotheses contain opposing viewpoints.

**$H_o$: The null hypothesis:** It is a statement about the population that either is believed to be true or is used to put forth an argument unless it can be shown to be incorrect beyond a reasonable doubt.

**$H_a$: The alternative hypothesis:** It is a claim about the population that is contradictory to H0 and what we conclude when we reject H0.

Since the null and alternative hypotheses are contradictory, you must examine evidence to decide if you have enough evidence to reject the null hypothesis or not. The evidence is in the form of sample data.

After you have determined which hypothesis the sample supports, you make adecision. There are two options for a decision. They are ‚Äúreject H0‚Äù if the sample information favors the alternative hypothesis or ‚Äúdo not reject H0‚Äù or ‚Äúdecline to reject H0‚Äù if the sample information is insufficient to reject the null hypothesis.

Mathematical Symbols Used in H0 and Ha:

|$H_0$	|$H_a$|
----|---|
|equal (=) |	not equal (‚â†) or greater than (>) or less than (<)|
|greater than or equal to (‚â•)|	less than (<)|
|less than or equal to (‚â§)	|more than (>)|

**Note:-**

$H_0$ always has a symbol with an equal in it. $H_a$ never has a symbol with an equal in it. The choice of symbol depends on the wording of the hypothesis test. However, be aware that many researchers (including one of the co-authors in research work) use = in the null hypothesis, even with > or < as the symbol in the alternative hypothesis. This practice is acceptable because we only make the decision to reject or not reject 
the null hypothesis.

<font color = "red">**Level of Significance & Hypothesis Testing**</font>

Significance levels in statistics are a crucial component of hypothesis testing. However, unlike other values in your statistical output, the significance level is not something that statistical software calculates. Instead, you choose the significance level. 

**Evidentiary Standards in the Courtroom:-**

Criminal cases and civil cases vary greatly, but they both require a minimum amount of evidence to convince a judge or jury to prove a claim against the defendant.

Prosecutors in criminal cases must prove the defendant is guilty ‚Äúbeyond a reasonable doubt,‚Äù whereas plaintiffs in a civil case must present a ‚Äúpreponderance of the evidence.‚Äù These terms are evidentiary standards that reflect the amount of evidence that civil and criminal cases require. These terms are evidentiary standards that reflect the amount of evidence that civil and criminal cases require.

For civil cases, most scholars define a preponderance of evidence as meaning that at least 51% of the evidence shown supports the plaintiff‚Äôs claim. However, criminal cases are more severe and require stronger evidence, which must go beyond a reasonable doubt. Most scholars define that evidentiary standard as being 90%, 95%, or even 99% sure that the defendant is guilty.

In statistics, the significance level is the evidentiary standard. For researchers to successfully make the case that the effect exists in the population, the sample must contain a sufficient amount of evidence.

In court cases, you have evidentiary standards because you don‚Äôt want to convict innocent people.

In hypothesis tests, we have the significance level because we don‚Äôt want to claim that an effect or relationship exists when it does not exist.

In statistics, the significance level defines the strength of evidence in probabilistic terms. Specifically, alpha represents the probability that tests will produce statistically significant results when the null hypothesis is correct. Rejecting a true null hypothesis is a type I error. And, the significance level equals the type I error rate. You can think of this error rate as the probability of a false positive. The test results lead you to believe that an effect exists when it actually does not exist.

Obviously, when the null hypothesis is correct, we want a low probability that hypothesis tests will produce statistically significant results. For example, if alpha is 0.05, your analysis has a 5% chance of producing a significant result when the null hypothesis is correct.

Just as the evidentiary standard varies by the type of court case, you can set the significance level for a hypothesis test depending on the consequences of a false positive. By changing alpha, you increase or decrease the amount of evidence you require in the sample to conclude that the effect exists in the population.

**It represents the probability of rejecting the null hypothesis when it is actually true, also known as Type 1 error. That means The significance level (Œ±) sets the probability of making a Type I error, which means rejecting a true null hypothesis. A 5% significance level means there's a 5% risk of making this mistake.**

<font color = "Red">**Rejection Region:-**</font>

A rejection region, also called a critical region, is an area where the null hypothesis is rejected. If the test results are significant enough to throw out the null hypothesis, they will fall into this rejection area. The primary purpose of statistics is to test theories or experimental results.

![alt text](images\Rejection_region.svg_.png)

**Problem wih Rejection Region**

Assume : Arbitrary Thresholds: If your Z-score is 1.95, you fail to reject H‚ÇÄ. But if it's 1.97, you reject it. The decision is binary‚Äîeven though 1.95 and 1.97 are almost identical, they lead to completely different conclusions.

Lack of Evidence Strength: The rejection region method only tells you whether you reject or fail to reject H‚ÇÄ. It does not tell you how strong the evidence is against H‚ÇÄ, unlike a p-value, which provides a more gradual measure of confidence.

Sensitivity to Sample Size: A larger sample size makes it easier to enter the rejection region, even if the real difference is tiny and practically not important. This could lead to rejecting H‚ÇÄ based on statistical significance, but not practical significance.

No Probabilistic Interpretation: P-values give a probabilistic measure of how extreme the test statistic is under H‚ÇÄ. The rejection region, however, just sets a fixed boundary, without giving any probability-based insight.

Using a p-value approach instead of just a rejection region would let you quantify the strength of evidence rather than making a strict reject/fail-to-reject decision.

<font size =12 color ="red">Types I & Type II Errors in Hypothesis Testing</font>

Hypothesis tests use sample data to make inferences about the properties of a population. You gain tremendous benefits by working with random samples because it is usually impossible to measure the entire population.

However, there are tradeoffs when you use samples. The samples we use are typically a minuscule percentage of the entire population. Consequently, they occasionally misrepresent the population severely enough to cause hypothesis tests to make Type I and Type II errors.

<font color = "red">**Potential outcomes in Hypothsis Testing**</font>

Hypothesis testing  is a procedure in inferential statistics that assesses two mutually exclusive theories about the properties of a population. For a generic hypothesis test, the two hypotheses are as follows:

        Null hypothesis: There is no effect

        Alternative hypothesis: There is an effect.

The sample data must provide sufficient evidence to reject the null hypothesis and conclude that the effect exists in the population. Ideally, a hypothesis test fails to reject the null hypothesis when the effect is not present in the population, and it rejects the null hypothesis when the effect exists.

Statisticians define two types of errors in hypothesis testing. Creatively, they call these errors Type I and Type II errors. Both types of error relate to incorrect conclusions about the null hypothesis.

![alt text](images\type-i-and-ii-errors1-l.jpg)

<font color = "Red" >**Fire Alarm Anology for types of errors**</font>

A fire alarm provides a good analogy for the types of hypothesis testing errors. Preferably, the alarm rings when there is a fire and does not ring in the absence of a fire. However, if the alarm rings when there is no fire, it is a false positive, or a Type I error in statistical terms. Conversely, if the fire alarm fails to ring when there is a fire, it is a false negative, or a Type II error.

<font color = "Red" >**Type-I (False Positive)**</font> error occurs when the sample results lead to the rejection of the null hypothesis when it is in fact true.

In other words, it's the mistake of finding a significant effect or relationship when there is none. The probability of committing a Type I error is denoted by Œ± (alpha), which is also known as the significance level. By choosing a significance level, researchers can control the risk of making a Type I error.

<font color = "Red" >**Type-II (False Negative)**</font>error occurs when based on the sample results, the null hypothesis is not rejected when it is in fact false.This means that the researcher fails to detect a significant effect or relationship when one actually exists. The probability of committing a Type II error is denoted by Œ≤ (beta). Trade-off between Type 1 and Type 2 errors

<font color = "red">**One-Tailed Test (One-Sided)**</font>

A one-tailed test is used when the researcher is interested in testing the effect in a **specific direction** (either greater than or less than the value stated in the null hypothesis).  
- The **alternative hypothesis** contains an inequality (`>` or `<`).

**Example:**  
A researcher wants to test whether a new medication **increases** the average recovery rate compared to the existing medication.  
- Null Hypothesis (H‚ÇÄ): Œº ‚â§ Œº‚ÇÄ  
- Alternative Hypothesis (H‚ÇÅ): Œº > Œº‚ÇÄ  



<font color = "red"> **Two-Tailed Test (Two-Sided)**</font>

A two-tailed test is used when the researcher wants to check if the value is **different** from the null hypothesis, in **either direction** (greater or lesser).  
- The **alternative hypothesis** uses the "not equal to" sign (`‚â†`).

**Example:**  
A researcher wants to test whether a new medication has a **different** average recovery rate (could be higher or lower) than the existing medication.  
- Null Hypothesis (H‚ÇÄ): Œº = Œº‚ÇÄ  
- Alternative Hypothesis (H‚ÇÅ): Œº ‚â† Œº‚ÇÄ  



## Advantages and Disadvantages

### **Two-Tailed Test**

#### ‚úÖ Advantages:

1. **Detects effects in both directions**  
   Useful when we are unsure about the direction of the effect.  
   _Example:_ A teacher wants to test if a new teaching method leads to a **change** in test scores (either improvement or decline).

2. **More conservative (lower Type I error risk)**  
   Since Œ± is split into two tails, there's a lower risk of wrongly rejecting the null hypothesis when it's true.  
   _Example:_ A pharmaceutical company uses two-tailed tests to be extra cautious before declaring a new drug is better or worse.

#### ‚ùå Disadvantages:

1. **Less powerful**  
   Because Œ± is divided between two tails, it may need a stronger signal (effect size) to reject H‚ÇÄ.  
   _Example:_ A small improvement in recovery rate may not be detected if we use a two-tailed test, even if it's real.

2. **Not suitable for directional hypotheses**  
   When your research question is clearly one-directional, two-tailed tests might be overkill.  
   _Example:_ If you're only interested in **improving** test scores, testing for both directions isn‚Äôt necessary.


### **One-Tailed Test**

#### ‚úÖ Advantages:

1. **More powerful**  
   Entire Œ± is in one tail, increasing the ability to detect a real effect in the expected direction.  
   _Example:_ A manager wants to check if a new process **increases** production speed. A one-tailed test would more easily show significant improvement.

2. **Good for directional hypotheses**  
   Ideal when theory or logic strongly supports checking for effect in only one direction.  
   _Example:_ A scientist tests if a fertilizer **increases** plant growth based on prior studies.

#### ‚ùå Disadvantages:

1. **Missed effects in opposite direction**  
   If the effect goes in the opposite direction, it won't be detected.  
   _Example:_ You test whether a drug improves health, but it actually makes it worse ‚Äî a one-tailed test won‚Äôt catch that!

2. **Higher risk of Type I error (if misused)**  
   If you wrongly assume the direction and use one-tailed testing, you might conclude there‚Äôs an effect when it's just random variation.  
   _Example:_ Choosing a one-tailed test just to get significant results (p-hacking) can lead to wrong scientific conclusions.

<font color = "red" size =12>**P-Value**</font>

A p-value is the probability of getting a result as extreme as or more extreme than the one you got, assuming the null hypothesis is true.

So, it‚Äôs not the probability that your null hypothesis is true.

**üéØ Example to Understand Intuition::-**

Let‚Äôs say you're testing whether a coin is fair.

Null Hypothesis (H‚ÇÄ): The coin is fair (P = 0.5 for heads).

You toss it 10 times and get 9 heads.

Now you ask: If the coin were really fair, what's the chance of getting 9 or more heads in 10 tosses?

‚û°Ô∏è That chance is your p-value.

If that p-value is very small (say 0.01 or 1%), it tells you: Whoa, that‚Äôs unlikely if the coin were fair! So maybe the coin isn‚Äôt fair.

**Example**
Suppose a company is evaluating the impact of a new training program on the productivity of its employees. The
company has data on the average productivity of its employees before implementing the training program. The
average productivity was 50 units per day. After implementing the training program, the company measures the
productivity of a random sample of 30 employees. The sample has an average productivity of 53 units per day and
the pop std is 4. The company wants to know if the new training program has significantly increased productivity.


Solution:- 

---

üìå Problem Recap:

* **Before training:** Average productivity = 50 units/day (Œº‚ÇÄ)  
* **After training:** Sample mean = 53 units/day (xÃÑ)  
* **Population standard deviation (œÉ):** 4  
* **Sample size (n):** 30  
* **Goal:** Has productivity **significantly increased** after training?

This is a **right-tailed hypothesis test**, since we‚Äôre testing for an **increase**.


‚úÖ Step-by-Step Solution:

**Step 1: State the Hypotheses**

* **H‚ÇÄ (Null Hypothesis):** Œº = 50  
* **H‚ÇÅ (Alternative Hypothesis):** Œº > 50 (one-tailed test)



**Step 2: Compute the Standard Error (SE)**

$$
SE = \frac{\sigma}{\sqrt{n}} = \frac{4}{\sqrt{30}} ‚âà 0.730
$$



**Step 3: Calculate the Z-score**

$$
Z = \frac{\bar{x} - \mu_0}{SE} = \frac{53 - 50}{0.730} ‚âà 4.11
$$

This means the sample mean (53) is 4.11 standard errors **above** the hypothesized mean (50).


**Step 4: Find the p-value**

This is the probability of getting a Z value **greater than 4.11** (since it's a right-tailed test):

$$
p = P(Z > 4.11) ‚âà 0.00001996
$$


**Step 5: Compare p-value with Œ± (significance level)**

* Common Œ± = 0.05  
* Our p-value ‚âà **0.00002** ‚Üí way **less than 0.05**



**‚úÖ Conclusion:**

Since **p-value < Œ±**, we **reject the null hypothesis**.

> The training program **significantly increased** employee productivity.

## üß™ Hypothesis Testing Using Z-Test (Two-Tailed)

**üìù Problem Setup**

A consumer watchdog organization wants to test whether the actual weight **differs significantly** from this claim.



**üìä Sample Data Given**

- Population mean (Œº) = **50 grams**
- Sample mean ($\bar{x}$) = **49 grams**
- Population standard deviation (œÉ) = **5 grams**
- Sample size (n) = **40**


**Step 1: Define the Hypotheses**

This is a **two-tailed** hypothesis test.

- **Null Hypothesis (H‚ÇÄ):** Œº = 50  
- **Alternative Hypothesis (H‚ÇÅ):** Œº ‚â† 50



**Step 2: Compute the Z-Score**

$$
Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}} = \frac{49 - 50}{\frac{5}{\sqrt{40}}} = \frac{-1}{0.79} ‚âà -1.27
$$



**Step 3: Find the p-value**

Using the Z-table:

- Area to the **left of z = -1.27** ‚âà **0.1020**
- Since it's a **two-tailed test**:

$$
\text{p-value} = 2 \times P(Z < -1.27) = 2 \times 0.1020 = 0.204
$$



**Step 4: Compare p-value with Significance Level (Œ±)**

- Let‚Äôs take Œ± = **0.05**
- p-value = **0.204** > **0.05**



‚úÖ **Conclusion**

Since the **p-value > Œ±**, we **fail to reject the null hypothesis**.

> üìå **Interpretation**:  
There is **not enough statistical evidence** to claim that the actual mean weight of Lays packets is different from 50 grams.  
The observed difference (49g vs 50g) could be due to **random chance**.

<font color ="red"  size = 12>**T-Test**</font>

A t-test is a statistical test used in hypothesis testing to compare the means of two samples or to compare a sample mean to a known population mean. The t-test is based on the t-distribution, which is used when the population standard deviation is unknown and the sample size is small.

There are three main types of t-tests:-

**One-sample t-test:** The one-sample t-test is used to compare the mean of a single sample to a known population mean. The null hypothesis states that there is no significant difference between the sample mean and the population mean, while the alternative hypothesis states that there is a significant difference.

**Independent two-sample t-test:** The independent two-sample t-test is used to compare the means of two independent samples. The null hypothesis states that there is no significant difference between the means of the two samples, while the alternative hypothesis states that there is a significant difference.

Example :- We take sample from two two section of same class.

**Paired t-test (dependent two-sample t-test):** The paired t-test is used to compare the means of two samples that are dependent or paired, such as pre-test and post-test scores for the same group of subjects or measurements taken on the same subjects under two different conditions. The null hypothesis states that there is no significant difference between the means of the paired differences, while the alternative hypothesis states that there is a significant difference.

Example:- Emplpyees Performance increase or decrease after some test.

<font color = "Red">**One-sample t-test:**</font>

A one-sample t-test checks whether a sample mean differs from the population mean.

Assumptions for a single sample t-test:-
1. Normality - Population from which the sample is drawn is normally distributed.

2. Independence - The observations in the sample must be independent, which means that the value of one observation should not influence the value of another observation.

3. Random Sampling - The sample must be a random and representative subset of thepopulation.

4. Unknown population std - The population std is not known.

**Case Study Example of One sample T-test:-**



Suppose a manufacturer claims that the average weight of their new chocolate bars is **50 grams**.  
We want to test this claim based on a **sample of 25 chocolate bars**, which had:

- Sample Mean (ùë•ÃÑ) = **49.7 grams**  
- Sample Standard Deviation (s) = **1.2 grams**  
- Sample Size (n) = **25**  
- Significance Level (Œ±) = **0.05**



**‚úÖ Step 1: Set the Hypotheses**

- **Null Hypothesis (H‚ÇÄ):** Œº = 50  
- **Alternative Hypothesis (H‚ÇÅ):** Œº ‚â† 50 (two-tailed)


**‚úÖ Step 2: Calculate Standard Error (SE)**

$$
SE = \frac{s}{\sqrt{n}} = \frac{1.2}{\sqrt{25}} = \frac{1.2}{5} = 0.24
$$


**‚úÖ Step 3: Compute the T-Statistic**

$$
t = \frac{\bar{x} - \mu}{SE} = \frac{49.7 - 50}{0.24} = \frac{-0.3}{0.24} ‚âà -1.25
$$



**‚úÖ Step 4: Degrees of Freedom**

$$
df = n - 1 = 25 - 1 = 24
$$



**‚úÖ Step 5: Determine P-Value**

From the t-distribution table or calculator:

- **t = -1.25**, **df = 24**
- **Two-tailed p-value ‚âà 0.223**


**‚úÖ Step 6: Conclusion**

Compare the p-value with Œ±:

- **p = 0.223 > Œ± = 0.05**

‚õî Since the p-value is **greater than Œ±**, we **fail to reject the null hypothesis**.


**üßæ Final Conclusion:**

There is **not enough evidence** to suggest that the mean weight of the chocolate bars differs significantly from 50 grams.  
‚û°Ô∏è The manufacturer's claim **is not contradicted by the sample**.

<font color = "red">**Independent Two sample T-test**</font>

An independent two-sample t-test, also known as an unpaired t-test, is a statistical method
used to compare the means of two independent groups to determine if there is a significant
difference between them.

Assumptions for the test:

1. Independence of observations: The two samples must be independent, meaning there is
no relationship between the observations in one group and the observations in the other
group. The subjects in the two groups should be selected randomly and independently.


2. Normality: The data in each of the two groups should be approximately normally
distributed. The t-test is considered robust to mild violations of normality, especially
when the sample sizes are large (typically n ‚â• 30) and the sample sizes of the two groups
are similar. If the data is highly skewed or has substantial outliers, consider using a non-
parametric test, such as the Mann-Whitney U test.


3. Equal variances (Homoscedasticity): The variances of the two populations should be
approximately equal. This assumption can be checked using F-test for equality of
variances. If this assumption is not met, you can use Welch's t-test, which does not
require equal variances.


4. Random sampling: The data should be collected using a random sampling method from
the respective populations. This ensures that the sample is representative of the
population and reduces the risk of selection bias.

**Example:-**

Suppose a website owner claims that there is no difference in the average time spent on their
website between desktop and mobile users. To test this claim, we collect data from 30
desktop users and 30 mobile users regarding the time spent on the website in minutes. The
sample statistics are as follows:

desktop users = [12, 15, 18, 16, 20, 17, 14, 22, 19, 21, 23, 18, 25, 17, 16, 24, 20, 19, 22, 18, 15,
14, 23, 16, 12, 21, 19, 17, 20, 14]

mobile_users = [10, 12, 14, 13, 16, 15, 11, 17, 14, 16, 18, 14, 20, 15, 14, 19, 16, 15, 17, 14, 12,
11, 18, 15, 10, 16, 15, 13, 16, 11]

Desktop users:

> Sample size (n1): 30

> Sample mean (mean1): 18.5 minutes

> Sample standard deviation (std_dev1): 3.5 minutes

Mobile users:

> Sample size (n2): 30

> Sample mean (mean2): 14.3 minutes

> Sample standard deviation (std_dev2): 2.7 minutes


We will use a significance level (Œ±) of 0.05 for the hypothesis test.


****Hypothesis Testing for Average Time Spent on Website**

**Step 1: Define Hypotheses**

Null hypothesis ($H_0$): There is no difference in the average time spent between desktop and mobile users.


$$
H_0: \mu_1 = \mu_2
$$



Alternative hypothesis ($H_A$): There is a difference in the average time spent.


$$
H_A: \mu_1 \neq \mu_2
$$



Where:
- $ \mu_1 $ is the mean time spent by desktop users.
- $ \mu_2 $ is the mean time spent by mobile users.

### Step 2: Calculate Test Statistic

The t-test statistic for independent samples with equal variance is given by:


$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{s_p^2 \left(\frac{1}{n_1} + \frac{1}{n_2} \right)}}
$$



Where:
- $ \bar{x}_1 = 18.5 $, $ \bar{x}_2 = 14.3 $ (sample means)
- $ s_1 = 3.5 $, $ s_2 = 2.7 $ (sample standard deviations)
- $ n_1 = 30 $, $ n_2 = 30 $ (sample sizes)
- **Pooled standard deviation** ($s_p$) is:



$$
s_p^2 = \frac{(n_1 - 1) s_1^2 + (n_2 - 1) s_2^2}{n_1 + n_2 - 2}
$$



Substituting values:



$$
s_p^2 = \frac{(30-1)(3.5^2) + (30-1)(2.7^2)}{30 + 30 - 2}
$$





$$
= \frac{(29)(12.25) + (29)(7.29)}{58}
$$




$$
= \frac{355.25 + 211.41}{58} = \frac{566.66}{58} = 9.77
$$



Taking square root:



$$
s_p = \sqrt{9.77} = 3.13
$$



Now, compute $t$:



$$
t = \frac{18.5 - 14.3}{\sqrt{9.77 \times \left(\frac{1}{30} + \frac{1}{30} \right)}}
$$





$$
= \frac{4.2}{\sqrt{9.77 \times 0.0667}}
$$





$$
= \frac{4.2}{\sqrt{0.651}} = \frac{4.2}{0.807} = 5.20
$$



### Step 3: Determine Critical Value

Using a **two-tailed t-test** at \( \alpha = 0.05 \) with degrees of freedom:



$$
df = n_1 + n_2 - 2 = 30 + 30 - 2 = 58
$$



From t-distribution tables, the critical value $t_{\alpha/2, df} $ for $ df = 58 $ at $ \alpha = 0.05 $ is **approximately 2.00**.

**Step 4: Compute the P-value**

The p-value represents the probability of obtaining a test statistic at least as extreme as the one computed, assuming that the null hypothesis is true. Since we have a two-tailed test, we need to find:



$$
P(T \geq |5.20|)
$$



Using a t-distribution table or a statistical calculator for $ t = 5.20 $ with $ df = 58 $, we find:



$$
p \approx 0.00001
$$



**Step 5: Decision Based on P-value**

The significance level is $ \alpha = 0.05 $.

Since $ p < 0.05 $, we **reject** the null hypothesis.

**Final Conclusion:-**

Since the p-value is extremely small, it provides **strong evidence** that the average time spent on the website is **significantly different** between desktop and mobile users. The website owner's claim **does not hold** statistically.

<font color = "red">**Paired Two sample t-twst**</font>

A paired two-sample t-test, also known as a dependent or paired-samples t-test, is a statistical
test used to compare the means of two related or dependent groups.

Common scenarios where a paired two-sample t-test is used include:

1. Before-and-after studies: Comparing the performance of a group before and after an
intervention or treatment.

2. Matched or correlated groups: Comparing the performance of two groups that are
matched or correlated in some way, such as siblings or pairs of individuals with similar
characteristics.


Assumptions

1. Paired observations: The two sets of observations must be related or paired in some way,
such as before-and-after measurements on the same subjects or observations from
matched or correlated groups.


2. Normality: The differences between the paired observations should be approximately
normally distributed. This assumption can be checked using graphical methods (e.g.,
histograms, Q-Q plots) or statistical tests for normality (e.g., Shapiro-Wilk test). Note that
the t-test is generally robust to moderate violations of this assumption when the sample
size is large.


3.  Independence of pairs: Each pair of observations should be independent of other pairs. In
other words, the outcome of one pair should not affect the outcome of another pair. This
assumption is generally satisfied by appropriate study design and random sampling.


**Paired Samples t-Test for Weight Loss Program**
Let's assume that a fitness center is evaluating the effectiveness of a new 8-week weight loss
program. They enroll 15 participants in the program and measure their weights before and
after the program. The goal is to test whether the new weight loss program leads to a
significant reduction in the participants' weight.

Before the program:

[80, 92, 75, 68, 85, 78, 73, 90, 70, 88, 76, 84, 82, 77, 91]
After the program:

[78, 93, 81, 67, 88, 76, 74, 91, 69, 88, 77, 81, 80, 79, 88]

Significance level (Œ±) = 0.05

**Step 1: Define Hypotheses**

Null hypothesis ($H_0$): The weight loss program **has no significant effect** on participants' weight.

$H_0: \mu_{\text{before}} = \mu_{\text{after}}$

Alternative hypothesis ($H_A$): The weight loss program **leads to a significant reduction** in participants' weight.

$H_A: \mu_{\text{before}} > \mu_{\text{after}}$

**Step 2: Compute Differences**

For a paired samples t-test, we compute the differences for each participant:

$$
d_i = x_{\text{before}, i} - x_{\text{after}, i}
$$



$$
\begin{array}{|c|c|c|c|}
\hline
\text{Participant} & \text{Before} & \text{After} & \text{Difference} (d_i) \\
\hline
1 & 80 & 78 & 2 \\
2 & 92 & 93 & -1 \\
3 & 75 & 81 & -6 \\
4 & 68 & 67 & 1 \\
5 & 85 & 88 & -3 \\
6 & 78 & 76 & 2 \\
7 & 73 & 74 & -1 \\
8 & 90 & 91 & -1 \\
9 & 70 & 69 & 1 \\
10 & 88 & 88 & 0 \\
11 & 76 & 77 & -1 \\
12 & 84 & 81 & 3 \\
13 & 82 & 80 & 2 \\
14 & 77 & 79 & -2 \\
15 & 91 & 88 & 3 \\
\hline
\end{array}
$$



**Step 3: Compute Test Statistic**

The formula for the paired t-test statistic is:

$$t = \frac{\bar{d}}{s_d / \sqrt{n}}$$

where:
- $ \bar{d} $ = Mean of differences
- $ s_d $ = Standard deviation of differences
- $ n $ = Number of pairs

**Compute $ \bar{d} $:**

$$
\bar{d} = \frac{\sum d_i}{n}
$$

$$
= \frac{2 -1 -6 +1 -3 +2 -1 -1 +1 +0 -1 +3 +2 -2 +3}{15}
$$

$$
= \frac{2}{15} = 0.13
$$

**Compute $ s_d $:**

$$
s_d = \sqrt{\frac{\sum (d_i - \bar{d})^2}{n - 1}}
$$

After calculation, we get:

$$
s_d = 2.45
$$

**Compute $ t $-value:**

$$
t = \frac{0.13}{2.45 / \sqrt{15}}
$$

$$
= \frac{0.13}{0.633} = 0.21
$$

**Step 4: Compute p-value**

Using a t-table or calculator, we find the **p-value** for $ t = 0.21 $ with **$ df = n - 1 = 15 - 1 = 14 $**.

$$
p \approx 0.42
$$

**Step 5: Conclusion**

Since $ p = 0.42 > 0.05 $, **we fail to reject the null hypothesis**.

Thus, there is **no significant evidence** to suggest that the weight loss program reduces participants' weight.
